```python
uhf = data[data['Band'] == 'VHF']
other = data[data['Band'] != 'VHF']
```

```python
uhf_mean = np.mean(uhf['Status_numeric'])
other_mean = np.mean(other['Status_numeric'])
```

```python
uhf_se = np.std(uhf['Status_numeric']) / len(uhf['Status_numeric'])**0.5
other_se = np.std(other['Status_numeric']) / len(other['Status_numeric'])**0.5
```

```python
uhf_dist = stats.t(df = len(uhf['Status_numeric'])-1, loc = uhf_mean, scale = uhf_se)
other_dist = stats.t(df = len(other['Status_numeric'])-1, loc = other_mean, scale = other_se)
```

```python
def get_t_dist(data):
    mean = np.mean(data)
    se = np.std(data) / len(data)**0.5
    return stats.t(df = len(data)-1, loc = mean, scale = se)
```

```python
def get_crit(dist, alpha=0.05, tail='two-sided'):
    if tail =='two-sided':
        lower = alpha/2
        upper = 1 - alpha/2
        return [dist.ppf(lower),dist.ppf(upper)]
    elif tail == 'greater':
        upper = 1 - alpha
        return [dist.ppf(upper)]
    elif tail == 'less':
        return [dist.ppf(alpha)]
```

```python
alpha = 0.05
fig, ax = plt.subplots(figsize=(15, 10))
x1 = np.linspace(uhf_dist.ppf(0.0001),uhf_dist.ppf(0.9999), 2000)
x2 = np.linspace(other_dist.ppf(0.0001),other_dist.ppf(0.9999), 2000)
ax.plot(x1, uhf_dist.pdf(x1), label='UHF');
[ax.axvline(x, color='black', linestyle='--') for x in get_crit(uhf_dist, alpha, tail='two-sided')]
ax.plot(x2, other_dist.pdf(x2), label='Others');
ax.legend(loc='best')
ax.set_title("Comparison Test")
fig.show()
fig.savefig('test.png', format='png')
```

```python
def create_exclusive_histogram(data, column, target, alpha = 0.05, tail='two-sided'):
    
    data_test =data[data[column] == target]
    data_other = data[data[column] != target]
    
    dist_test = get_t_dist(data_test)
    dist_other = get_t_dist(dist_other)
    
    fig, ax = plt.subplots(figsize=(15, 10))
    x1 = np.linspace(dist_test.ppf(0.0001),dist_test.ppf(0.9999), 2000)
    x2 = np.linspace(other_dist.ppf(0.0001),other_dist.ppf(0.9999), 2000)
    
    ax.plot(x1, dist_test.pdf(x1), label=target);
    
    [ax.axvline(x, color='black', linestyle='--') for x in get_crit(dist_test, alpha, tail)]
    
    ax.plot(x2, dist_other.pdf(x2), label=f'Non {target}');
    ax.legend(loc='best')
    ax.set_title(f"Distribution of {target} against sample")
    fig.show()
    fig.savefig('test.png', format='png')
```
